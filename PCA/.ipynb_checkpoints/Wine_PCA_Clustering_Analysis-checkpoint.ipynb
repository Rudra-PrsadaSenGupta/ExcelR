# PCA and Clustering Analysis

This notebook performs EDA, PCA, and Clustering analysis using the provided dataset.

## Task 1: Exploratory Data Analysis (EDA)

```python

# Load the dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('wine.csv')
data.head()
```

### Feature Distributions: Histograms and Boxplots

```python

# Plot histograms of the features
data.hist(figsize=(12, 12))
plt.show()

# Boxplot to check for outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.show()
```

### Correlation Matrix Heatmap

```python

# Compute correlation matrix and visualize with a heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## Task 2: Dimensionality Reduction with PCA

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the data (mean 0, std 1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply PCA
pca = PCA()
pca.fit(scaled_data)

# Plot explained variance ratio
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

### Transform the Data Using Principal Components

```python

# Set number of components (e.g., 2 for visualization)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Create a DataFrame with the PCA components
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_df.head()
```

## Task 3: Clustering with Original Data

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Apply K-Means clustering on the original dataset
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)

# Evaluate clustering performance with silhouette score
silhouette_original = silhouette_score(scaled_data, kmeans_labels)
print(f'Silhouette Score (Original Data): {silhouette_original}')

# Visualize the clusters
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering on Original Data')
plt.show()
```

## Task 4: Clustering with PCA Data

```python

# Apply K-Means clustering on the PCA-transformed dataset
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca_labels = kmeans_pca.fit_predict(pca_data)

# Evaluate clustering performance with silhouette score on PCA data
silhouette_pca = silhouette_score(pca_data, kmeans_pca_labels)
print(f'Silhouette Score (PCA Data): {silhouette_pca}')

# Visualize the clusters based on PCA components
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=kmeans_pca_labels, cmap='plasma')
plt.title('K-Means Clustering on PCA-Transformed Data')
plt.show()
```
