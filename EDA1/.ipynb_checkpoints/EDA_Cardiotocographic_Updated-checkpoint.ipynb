# Exploratory Data Analysis (EDA) - Cardiotocographic Dataset
This notebook contains steps to clean the dataset, handle missing values, remove duplicates, and visualize key relationships.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
df = pd.read_csv('Cardiotocographic.csv')
# Handling missing values and duplicates
df.fillna(df.median(), inplace=True)
df_no_duplicates = df.drop_duplicates()
```

## Data Visualization

```python
# Histograms for FHR, FM, UC
plt.figure(figsize=(16, 10))
plt.subplot(2, 2, 1)
sns.histplot(df_no_duplicates['LB'], kde=True)
plt.title('Distribution of Baseline FHR (LB)')
plt.subplot(2, 2, 2)
sns.histplot(df_no_duplicates['FM'], kde=True)
plt.title('Distribution of Fetal Movements (FM)')
plt.subplot(2, 2, 3)
sns.histplot(df_no_duplicates['UC'], kde=True)
plt.title('Distribution of Uterine Contractions (UC)')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(14,10))
correlation_matrix = df_no_duplicates.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Variables')
plt.show()
```
