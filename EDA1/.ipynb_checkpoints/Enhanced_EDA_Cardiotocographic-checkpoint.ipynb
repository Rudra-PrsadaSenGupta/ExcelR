# Exploratory Data Analysis on Cardiotocographic Dataset

```python

# Loading the dataset
import pandas as pd
df = pd.read_csv("Cardiotocographic.csv")

# Basic information about the dataset
df.info()
```

```python

# Handle missing values (using median imputation here)
df_cleaned = df.fillna(df.median())

# Check for and remove duplicates
df_cleaned = df_cleaned.drop_duplicates()

# Check the dataset after cleaning
df_cleaned.info()
```

## Data Cleaning and Preparation

```python

# Advanced Visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplots for selected features
plt.figure(figsize=(16, 10))
plt.subplot(2, 2, 1)
sns.boxplot(df_cleaned['LB'])
plt.title('Boxplot of Baseline FHR (LB)')

plt.subplot(2, 2, 2)
sns.boxplot(df_cleaned['FM'])
plt.title('Boxplot of Fetal Movements (FM)')

plt.subplot(2, 2, 3)
sns.boxplot(df_cleaned['UC'])
plt.title('Boxplot of Uterine Contractions (UC)')

plt.tight_layout()
plt.show()

# Scatter plot to see relationships
sns.pairplot(df_cleaned[['LB', 'FM', 'UC', 'ASTV', 'MLTV']])
plt.show()
```

## Statistical Summary

```python

# Statistical summary of the dataset
df.describe()
```

## Data Visualization

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the distribution of numerical columns
numerical_cols = df.select_dtypes(include='number').columns.tolist()

# Plotting histograms for each numerical column
df[numerical_cols].hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
