# Enhanced EDA - Cardiotocographic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Cardiotocographic.csv')
df_cleaned = df.fillna(df.median())
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Histograms and Boxplots for distribution analysis
plt.figure(figsize=(16, 10))
plt.subplot(2, 2, 1)
sns.histplot(df_cleaned['LB'], kde=True)
plt.title('Distribution of Baseline FHR (LB)')
plt.subplot(2, 2, 2)
sns.boxplot(df_cleaned['LB'])
plt.title('Boxplot of Baseline FHR (LB)')
plt.subplot(2, 2, 3)
sns.histplot(df_cleaned['FM'], kde=True)
plt.title('Distribution of Fetal Movements (FM)')
plt.subplot(2, 2, 4)
sns.boxplot(df_cleaned['FM'])
plt.title('Boxplot of Fetal Movements (FM)')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap of cleaned data
plt.figure(figsize=(14,10))
correlation_matrix_cleaned = df_cleaned.corr()
sns.heatmap(correlation_matrix_cleaned, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Cleaned Variables')
plt.show()
```
