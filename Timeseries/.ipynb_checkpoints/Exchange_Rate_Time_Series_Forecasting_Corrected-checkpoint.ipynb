# Forecasting Exchange Rates using Time Series Analysis
This notebook demonstrates the use of ARIMA and Exponential Smoothing techniques to forecast future exchange rates based on the provided dataset 'exchange_rate.csv'. The goal is to analyze the historical exchange rate data and build models for forecasting future trends.

## Step 1: Data Preparation and Exploration

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
exchange_rate_data = pd.read_csv('exchange_rate.csv')

# Check the column names to identify the date column
exchange_rate_data.columns
```

```python
# Parse the correct date column (assuming the first column is the date)
exchange_rate_data['Date'] = pd.to_datetime(exchange_rate_data.iloc[:, 0])

# Set 'Date' as the index and drop the original date column
exchange_rate_data.set_index('Date', inplace=True)
exchange_rate_data.drop(exchange_rate_data.columns[0], axis=1, inplace=True)

# Display the first few rows after setting the index
exchange_rate_data.head()
```

```python
# Plot the time series to understand trends
plt.figure(figsize=(10,6))
plt.plot(exchange_rate_data)
plt.title('Exchange Rate (USD to AUD)')
plt.ylabel('Exchange Rate')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

```python
# Check for missing values
print(exchange_rate_data.isnull().sum())

# Fill missing values with forward fill method
exchange_rate_data.fillna(method='ffill', inplace=True)

# Recheck for missing values
print("Missing values after fill:", exchange_rate_data.isnull().sum())
```

## Step 2: ARIMA Model - Parameter Estimation and Fitting

```python
# Import necessary libraries for ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# Plot ACF and PACF to determine p and q values
plt.figure(figsize=(10, 5))
plt.subplot(121)
plot_acf(exchange_rate_data, lags=30, ax=plt.gca())
plt.subplot(122)
plot_pacf(exchange_rate_data, lags=30, ax=plt.gca())
plt.show()
```

```python
# Fit the ARIMA model
arima_model = ARIMA(exchange_rate_data, order=(1, 1, 1))
arima_result = arima_model.fit()

# Summary of the ARIMA model
print(arima_result.summary())
```

```python
# Forecast future exchange rates
forecast = arima_result.forecast(steps=30)  # Forecast the next 30 days

# Plot the forecasted values against the actual values
plt.figure(figsize=(10, 6))
plt.plot(exchange_rate_data, label='Actual')
plt.plot(forecast, label='Forecast', linestyle='--', color='red')
plt.title('ARIMA Forecast vs Actual')
plt.legend()
plt.show()
```

## Step 3: Exponential Smoothing Model

```python
# Import necessary libraries for Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit an Exponential Smoothing model (Holt's Linear Trend)
exp_model = ExponentialSmoothing(exchange_rate_data, trend='add', seasonal=None)
exp_result = exp_model.fit()

# Forecast future values
exp_forecast = exp_result.forecast(steps=30)

# Plot the forecasted values vs actual values
plt.figure(figsize=(10, 6))
plt.plot(exchange_rate_data, label='Actual')
plt.plot(exp_forecast, label='Exponential Smoothing Forecast', linestyle='--', color='green')
plt.title('Exponential Smoothing Forecast vs Actual')
plt.legend()
plt.show()
```

## Step 4: Model Evaluation and Comparison

```python
# Define error metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate MAE, RMSE, and MAPE for ARIMA
mae_arima = mean_absolute_error(exchange_rate_data[-30:], forecast)
rmse_arima = np.sqrt(mean_squared_error(exchange_rate_data[-30:], forecast))
mape_arima = np.mean(np.abs((exchange_rate_data[-30:] - forecast) / exchange_rate_data[-30:])) * 100

# Calculate MAE, RMSE, and MAPE for Exponential Smoothing
mae_exp = mean_absolute_error(exchange_rate_data[-30:], exp_forecast)
rmse_exp = np.sqrt(mean_squared_error(exchange_rate_data[-30:], exp_forecast))
mape_exp = np.mean(np.abs((exchange_rate_data[-30:] - exp_forecast) / exchange_rate_data[-30:])) * 100

# Print the error metrics
print(f"ARIMA - MAE: {mae_arima:.3f}, RMSE: {rmse_arima:.3f}, MAPE: {mape_arima:.2f}%")
print(f"Exponential Smoothing - MAE: {mae_exp:.3f}, RMSE: {rmse_exp:.3f}, MAPE: {mape_exp:.2f}%")
```

## Conclusion
In this notebook, we used two time series forecasting techniques: ARIMA and Exponential Smoothing. Both models were evaluated using MAE, RMSE, and MAPE error metrics. Based on the results, we can compare the models and choose the one that provides the most accurate forecast for future exchange rates.
